```python
#imports

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("benchmark_results.csv")
workloads = df['Workload']
original_acc = df['Original Model Accuracy']
distilled_acc = df['Distilled Model Accuracy']


```

```python
# Bar Plot
plt.figure(figsize=(10, 5))
x = np.arange(len(workloads))
width = 0.35
plt.bar(x - width/2, original_acc, width, label='Original Model')
plt.bar(x + width/2, distilled_acc, width, label='Distilled Model')
plt.xticks(ticks=x, labels=workloads, rotation=45)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison Across Workloads")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Line Plot
plt.figure(figsize=(10, 5))
plt.plot(workloads, original_acc, marker='o', linestyle='-', label='Original Model', color='b')
plt.plot(workloads, distilled_acc, marker='s', linestyle='--', label='Distilled Model', color='r')
plt.ylabel("Accuracy")
plt.title("Accuracy Trend Across Workloads")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 5))
sns.scatterplot(x=list(workloads) * 2, y=list(original_acc) + list(distilled_acc), hue=["Original"]*len(workloads) + ["Distilled"]*len(workloads), style=["Original"]*len(workloads) + ["Distilled"]*len(workloads), s=100)
plt.ylabel("Accuracy")
plt.title("Scatter Plot of Model Performance")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
